# Explanation about each and every column

**column 1.Model**

Indicates the specific BMW model name, such as 5 Series or X3.<br>
Helps identify which model variants are most popular or highest-priced.

**column 2. Year**

Shows the manufacturing year of the vehicle.<br>
Newer models generally have higher prices and lower mileage.

**column 3. Region**

Specifies the geographical market where the car was sold.<br>
Useful for analyzing regional trends, pricing, and customer preferences.

**column 4. Color**

Represents the exterior color of the vehicle.<br>
Color can influence customer preference and resale value.

**column 5. Fuel_Type**

Indicates the type of fuel the car uses (Petrol, Diesel, Hybrid, Electric).<br>

**column 6. Transmission**

Transmission type influences driving comfort and vehicle pricing.

**column 7. Engine_Size_L**

Represents the engine capacity in liters (e.g., 2.0L, 3.5L).<br>
Higher engine size usually means better performance but lower mileage.

**column 8. Mileage_KM**

Indicates how many kilometers the car has been driven.<br>
Mileage affects the car’s condition and resale value.

**column 9. Price_USD**

Shows the selling price of the vehicle in US Dollars.<br>
Used to analyze pricing trends and build prediction models.

**column 10. Sales_Volume**

Reflects the total units sold for that model.<br>
Helps identify high-demand and low-demand cars.

**column 11. Sales_Classification**

Labels each record as having High or Low sales.<br>
Useful for segmentation and classification analysis.

```python
# --- 4. Outlier Detection (Numeric Columns) using IQR Method ---
print("\n=== Outlier Count (IQR method) ===")

for col in numeric_cols:
    Q1 = b[col].quantile(0.25)
    Q3 = b[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = b[(b[col] < lower) | (b[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
# --- 5. Check Inconsistent Formatting (extra spaces, mixed case) ---
print("\n=== Inconsistent Formatting Check ===")

for col in categorical_cols:
    trimmed = b[col].str.strip()
    if not trimmed.equals(b[col]):
        print(f"Inconsistent spacing found in column: {col}")

    if b[col].str.lower().nunique() != b[col].nunique():
        print(f"Case formatting inconsistency found in column: {col}")
```

 # Phase 2: Data Cleaning & Manipulation


**1 Handle Missing Data**
**Identify which columns have missing or null values.**
    
**Choose the appropriate strategy:**

**Remove rows or columns with too many missing entries.**

**Fill missing values using suitable methods such as:**

**Mean or median (for numerical values)**

**Mode or placeholder (for categorical values)**

**Justify the method chosen based on the type of data and its significance.**


```python
missing = b.isnull().sum().sort_values(ascending=False)
missing_percent = (b.isnull().sum() / len(b)) * 100
missing_summary = pd.DataFrame({
                        'Missing Count': missing, 
                        'Missing %': missing_percent
                        })

missing_summary[missing_summary['Missing Count'] > 0]

# for checking the percentage of missing of each column
```

```python
# --- Missing Values Summary ---
print(b.isnull().sum())

# --- Separate numeric and categorical columns ---
numeric_cols = b.select_dtypes(include=['int64', 'float64']).columns
category_cols = b.select_dtypes(include=['object']).columns

# --- 1. Fill numeric missing values with median ---
b[numeric_cols] = b[numeric_cols].fillna(b[numeric_cols].median()).round(2)

# --- 2. Fill categorical missing values with mode ---
for col in category_cols:
    b[col] = b[col].fillna(b[col].mode()[0])

# --- Check again ---
print("\nMissing values after handling:")
print(b.isnull().sum())
```
